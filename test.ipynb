```python
import pandas as pd

# Fetch documents from MongoDB
cursor = collection.find({}, {'_id': 1, 'host': 1})

# Convert cursor to list of dictionaries
host_data = list(cursor)

# Create DataFrame
df_host = pd.DataFrame(host_data)

# Extract keys from first document's 'host' dictionary
host_keys = list(df_host.iloc[0, 1].keys())
# Remove keys that are not needed
keys_to_remove = ['host_url','host_about', 'host_thumbnail_url', 'host_picture_url']
host_keys = [key for key in host_keys if key not in keys_to_remove]

# Extract host information into DataFrame columns
for key in host_keys:
    if key == 'host_response_time':
        df_host['host_response_time'] = df_host['host'].apply(lambda x: x.get('host_response_time', 'Not Specified'))
    else:
        df_host[key] = df_host['host'].apply(lambda x: x.get(key, 'Not Specified'))

# Drop the 'host' column
df_host.drop(columns=['host'], inplace=True)

# Display DataFrame
print(df_host.head())
```

```python
df_host['host_is_superhost'] = df_host['host_is_superhost'].map({False:'No',True:'Yes'})
df_host['host_has_profile_pic'] = df_host['host_has_profile_pic'].map({False:'No',True:'Yes'})
df_host['host_identity_verified'] = df_host['host_identity_verified'].map({False:'No',True:'Yes'})
df_host.head()
```

```python
import pandas as pd

# Fetch documents from MongoDB
cursor = collection.find({}, {'_id': 1, 'address': 1})

# Convert cursor to list of dictionaries
address_data = list(cursor)

# Create DataFrame
df_address = pd.DataFrame(address_data)

# Extract keys from first document's 'address' dictionary
address_keys = list(df_address.iloc[0, 1].keys())

# Extract address information into DataFrame columns
for key in address_keys:
    if key == 'location':
        df_address['location_type'] = df_address['address'].apply(lambda x: x.get('location', {}).get('type', 'Not Specified'))
        df_address['longitude'] = df_address['address'].apply(lambda x: x.get('location', {}).get('coordinates', [])[0])
        df_address['latitude'] = df_address['address'].apply(lambda x: x.get('location', {}).get('coordinates', [])[1])
        df_address['is_location_exact'] = df_address['address'].apply(lambda x: x.get('location', {}).get('is_location_exact', 'Not Specified'))
    else:
        df_address[key] = df_address['address'].apply(lambda x: x.get(key, 'Not Specified'))

# Drop the 'address' column
df_address.drop(columns=['address'], inplace=True)

# Display DataFrame
df_address.head()
```

```python
df_address['is_location_exact'] = df_address['is_location_exact'].map({False:'No',True:'Yes'})
df_address.head()
```

```python
availability = []
for i in collection.find( {}, {'_id':1, 'availability':1}):
    availability.append(i)

df_availability = pd.DataFrame(availability)
availability_keys = list(df_availability.iloc[0,1].keys())

for i in availability_keys:
    df_availability['availability_30'] = df_availability['availability'].apply(lambda x: x['availability_30'])
    df_availability['availability_60'] = df_availability['availability'].apply(lambda x: x['availability_60'])
    df_availability['availability_90'] = df_availability['availability'].apply(lambda x: x['availability_90'])
    df_availability['availability_365'] = df_availability['availability'].apply(lambda x: x['availability_365'])

df_availability.drop(columns=['availability'], inplace=True)
df_availability.head()
```

```python
df_availability.isnull().sum()
```

```python
def amenities_sort(x):
    a = x
    a.sort(reverse=False)
    return a

amenities = []
for i in collection.find( {}, {'_id':1, 'amenities':1}):
    amenities.append(i)

df_amenities = pd.DataFrame(amenities)
df_amenities['amenities'] = df_amenities['amenities'].apply(lambda x: amenities_sort(x))
df_amenities.head()
```

```python
df_amenities.isnull().sum()
```

```python
df = pd.merge(df, df_host, on='_id')
df = pd.merge(df, df_address, on='_id')
df = pd.merge(df, df_availability, on='_id')
df = pd.merge(df, df_amenities, on='_id')
df.head()
```
